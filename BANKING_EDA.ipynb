# Categorical Analysis

```python
# Examine the distribution of unique category in categorical columns
bins  = [0,100000,300000, float('inf')]
labels = ['Low','Mid','High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

# Univatriate Analysis

```python
categorical_cols = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for'{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

# Bivariate Analysis

```python
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor,hue="Nationality")
```

```python
# Histplot of value counts for differnt Occupation

for col in categorical_cols:
    if col == "Occuption":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# Numerical Analysis

```python
numerical_cols = ["Age","Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Saving Accounts","Business Lending"]

# Univariate Analysis

plt.figure(figsize=(15,20))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,2,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.tight_layout()    
plt.show()    
```

# Heatmaps

```python
numerical_cols = ["Age","Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Saving Accounts","Business Lending"]

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix, annot=True, cmap= 'crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

# Insights Of EDA

-> Strongest Positive Relationships: Bank Deposits and Saving Accounts share the highest correlation at 0.75. This indicates that clients with higher savings balances also tend to hold larger overall deposits.
-> Moderate Debt Correlations: Business Lending correlates positively with Bank Loans (0.42) and Bank Deposits (0.44). This suggests that businesses actively borrowing also maintain higher deposit accounts.
-> Income Drives Borrowing: Estimated Income shows a moderate positive relationship with Superannuation Savings (0.37), Credit Card Balance (0.30), and Business Lending (0.33). Higher earners tend to borrow more and save more for retirement.
